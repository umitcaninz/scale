<h4 style="color:purple;">Karekök dönüşümü nasıl yapılır? </h4>



\begin{equation}
\text{Karekök Dönüşümü} = \sqrt{x }
\end{equation}

<h1 style="color:purple;">Python İle Sağa Çarpık Veri Üzerinde Karekök Dönüşümü</h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Sağa çarpık (sağa çekik) bir dağılım oluşturmak için log-normal dağılım kullanacağız.

mean = 10  # Ortalama
sigma = 1  # Standart sapma

data = np.random.lognormal(mean, sigma, 100000)

# Aykırı değerler oluşturmak için, normal dağılım ile log-normal dağılımın birleşimi olacak şekilde bir veri seti oluşturacağız.

# Normal dağılım parametreleri
normal_mean = 30
normal_sigma = 10

normal_data = np.random.normal(normal_mean, normal_sigma, 5000)

data_with_outliers = np.concatenate((data, normal_data))

# Veriyi bir DataFrame'e dönüştür
df = pd.DataFrame(data_with_outliers, columns=['Değerler'])
```

```python
# Karekök dönüşümü
df['Karekök'] = df['Değerler'].apply(np.sqrt)
```

```python
# Orijinal veri dağılımı
plt.hist(df['Değerler'])
plt.xlabel("Değerler")
plt.ylabel("Sıklık")
plt.title("Orijinal Veri Dağılımı")
plt.show()

# Karekök dönüşümü sonrası veri dağılımı
plt.hist(df['Karekök'])
plt.xlabel("Karekök Değerleri")
plt.ylabel("Sıklık")
plt.title("Karekök Dönüşümü Sonrası Veri Dağılımı")
plt.show()
```

<h4 style="color:purple;">Çarpıklık Değerlerini Karşılaştırma: </h4>



```python
# Orijinal veri çarpıklığı
print("Orijinal Veri Çarpıklığı:", df['Değerler'].skew())

# Karekök dönüşümü sonrası veri çarpıklığı
print("Karekök Dönüşümü Sonrası Veri Çarpıklığı:", df['Karekök'].skew())
```

<h2 style="color:purple;">En Optimal Karekök Derecesi Belirleme:</h2>



```python
import pandas as pd
import numpy as np

# Karekök dereceleri
karekok_dereceleri = [0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5]

# Çarpıklık ve derece değerlerini saklamak için listeler
carpıklık_degerleri = []
karekok_dereceleri_list = []

# Min çarpıklık değerini ve derecesini saklamak için değişkenler
min_carpıklık = np.inf
min_carpıklık_derecesi = None

for power in karekok_dereceleri:
  # Karekök dönüşümü
  df[f'Karekök ({power})'] = df['Değerler'].apply(lambda x: x**(1/power))

  # Çarpıklık değeri
  carpıklık = df[f'Karekök ({power})'].skew()

  # Min çarpıklık değerini ve derecesini güncelleme
  if carpıklık < min_carpıklık:
    min_carpıklık = carpıklık
    min_carpıklık_derecesi = power

  # Listeleri doldurma
  carpıklık_degerleri.append(carpıklık)
  karekok_dereceleri_list.append(power)

# DataFrame oluşturma
df_carpıklık = pd.DataFrame({'Karekök Derecesi': karekok_dereceleri_list, 'Çarpıklık': carpıklık_degerleri})

# Min çarpıklık değerini ve derecesini yazdırma
print(f"Min Çarpıklık Değeri: {min_carpıklık}")
print(f"Min Çarpıklık Derecesi: {min_carpıklık_derecesi}")

# DataFrame'i yazdırma
df_carpıklık.sort_values(ascending=True,by="Çarpıklık")[:3]
```
